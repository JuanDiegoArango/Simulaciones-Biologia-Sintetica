```python
%pylab inline
from scipy.misc import factorial
```

```python
def poisson(k,g,n):
    return pow((k/g),n)*exp(-k/g)/factorial(n) 
```

```python
def aproximacion(k,g,n):
    return pow((g*n+k),k/g-1)*exp(-n/2)
    
```

para todas las gr√°ficas azul es aproximancion y rojo es poisson

tasa de creacion =10
tasa de destruccion=10

```python
k=10.0
g=10.0

x=linspace(1,200,1000)
n=arange(1,200)
scatter(n,poisson(k,g,n),color="red")
plot(n,poisson(k,g,n),color="red")
plot(x,aproximacion(k,g,x)/norm(aproximacion(k,g,x)))
xlim(0,20)  
ylabel("probabilidad")
xlabel("numero de moleculas")
```
